```python
import sagemaker
from sagemaker.core.helper.session_helper import Session
from sklearn.model_selection import train_test_split
import boto3
import pandas as pd
```

```python

sm_boto3 = boto3.client("sagemaker")
#sess = sagemaker.Session()
sess = Session()
region = sess.boto_session.region_name
bucket = "hariehkrmobsagemaker"
print("using bucket" +bucket)
```

```python
print(region)
```

```python
df=pd.read_csv("mob_price_classification_train.csv")
df.head()
```

```python
df['price_range'].value_counts()
```

```python
features=list(df.columns)
features
```

```python
label = features.pop(-1)
label
```

```python
x=df[features]
y=df[label]
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.15, random_state=0)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
trainX = pd.DataFrame(X_train)
trainX[label] = y_train

testX = pd.DataFrame(X_test)
testX[label] = y_test
```

```python
import os
os.makedirs("data")
```

```python
trainX.to_csv("data/train-V-1.csv",index = False)
testX.to_csv("data/test-V-1.csv", index = False)
```

```python
## send the data to s3. Sagemaker will take the data from training for s3
sk_prefix = "sagemaker/mobile_price_classification/sklearncontainer"
trainpath = sess.upload_data(path="data/train-V-1.csv", bucket=bucket, key_prefix=sk_prefix)
testpath = sess.upload_data(path="data/test-V-1.csv", bucket=bucket, key_prefix=sk_prefix)

print(trainpath)
print(testpath)
```
